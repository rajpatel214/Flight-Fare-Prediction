```python

air.fillna(air[(air["Source"]=="Delhi")&
                 (air["Destination"]=="Cochin")]["Total_Stops"].mode()[0],
            inplace=True)
```

```python
air.drop(["Route","Additional_Info"],axis=1,inplace=True)
```

```python
air
```

```python
import warnings
warnings.filterwarnings("ignore")

air["j_mon"]=pd.to_datetime(air["Date_of_Journey"]).dt.month
air["j_day"]=pd.to_datetime(air["Date_of_Journey"]).dt.day

air["d_hr"]=pd.to_datetime(air["Dep_Time"]).dt.hour
air["d_min"]=pd.to_datetime(air["Dep_Time"]).dt.minute

air["a_hr"]=pd.to_datetime(air["Arrival_Time"]).dt.hour
air["a_min"]=pd.to_datetime(air["Arrival_Time"]).dt.minute

```

```python
air=air.drop(["Date_of_Journey","Dep_Time","Arrival_Time"],axis=1)
```

```python
air
```

```python
lis=air["Duration"]

new_lis=[]
for i in lis:
    if len(i.split())==1:
        if "m" in i:
            i="0h "+i
        else:
            i=i+" 0m"
            print(i)
    new_lis.append(i)

air["Duration"]=new_lis

air["du_hr"]=air["Duration"].str.split(" ").str[0].replace("h","",regex=True).astype(int)
air["du_min"]=air["Duration"].str.split(" ").str[-1].replace("m","",regex=True).astype(int)

```

```python
air.drop("Duration",axis=1,inplace=True)
```

```python
air
```

```python
air["Total_Stops"].unique()

air["Total_Stops"].replace(['non-stop', '2 stops', '1 stop', '3 stops', '4 stops'],[0,2,1,3,4],inplace=True)

from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
air[["Airline","Source","Destination"]]=air[["Airline","Source","Destination"]].apply(enc.fit_transform)


x=air.drop("Price",axis=1)
y=air.Price

```

```python
from sklearn.model_selection import train_test_split 

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss1=StandardScaler()
```

```python
x_train_scling=ss.fit_transform(x_train)
```

```python
y_train_scling=ss1.fit_transform(pd.DataFrame(y_train))
```

```python
x_test_scling=ss.transform(x_test)
y_test_scling=ss1.transform(pd.DataFrame(y_test))
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
```

```python
model.fit(x_train_scling,y_train)
```

```python
model.score(x_train_scling,y_train)
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression

# from sklearn.linear_model import LogisticRegression
# from sklearn.ensemble import RandomForestClassifier
# from sklearn.svm import SVC

# Define base models
base_models = [
    ('rf', RandomForestRegressor(n_estimators=20, random_state=20)),
    ('svm', SVR())
]
# Define meta-model

stack_model = StackingRegressor(estimators=base_models, final_estimator=LinearRegression())
# Train the stacking model
stack_model.fit(x_train, y_train)
```

```python
stack_model.score(x_train,y_train)
```

```python
stack_model.score(x_test,y_test)
```
